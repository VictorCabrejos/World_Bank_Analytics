```python
# Define lists2dict()
def lists2dict(list1, list2):
    """Return a dictionary where list1 provides
    the keys and list2 provides the values."""

    # Zip lists: zipped_lists
    zipped_lists = zip(list1, list2)

    # Create a dictionary: rs_dict
    rs_dict = dict(zipped_lists)

    # Return the dictionary
    return rs_dict
```

```python
# Import the pandas package
import pandas as pd

# Turn list of lists into list of dicts: list_of_dicts
list_of_dicts = [lists2dict(names, sublist) for sublist in df1.values]

# Turn list of dicts into a dataframe: df2
df2 = pd.DataFrame(list_of_dicts)
print(df2.shape)

# Turn the first 2900 rows into df3
df3 = df2.iloc[0:2900,:] 

# Print df3
df3
```

## 4) Processing data in chunks 

* Processing a file line by line for a given number of lines (1000)
* Counting the number of ocurrences of the first column value

```python
# Open connection to file 'world_dev_ind.csv'
with open('world_dev_ind.csv') as file:
    
    # Skipe the column names
    file.readline()
    
    # Initialize an empty dictionary
    counts_dict = {}
    
    # Process only the first 1000 rows
    for i in range(0,1000):
        
        # Split the current line into a list
        line = file.readline().split(',')
        
        # Get the value for the first column
        first_col = line[0]
        
        # If the column value is in the dict, increment its value
        if first_col in counts_dict.keys():
            counts_dict[first_col] += 1
        else:
            counts_dict[first_col] = 1
            
for key, value in counts_dict.items():
    print("{} : {} times".format(key,value))
```

### * Writing a generator function to load data in chunks

* We will write a `Generator` function to process the entire file this time
* `'Lazy Evaluation'` is useful when dealing with `very large datasets`
* It lets you generate values in an efficient manner by `yielding` only `chunks` of data at a time

```python
def read_large_file(file_object):
    """ 
    A generator function to read a
    large file lazily 
    """
    
    # Loop indefinitely until the end of the file
    while True:
        
        # Read a line from the file
        data = file_object.readline()
        
        # Break if this is the end of the file
        if not data:
            break
            
        # Yield the line of data
        yield data
    
```

```python
with open('world_dev_ind.csv') as file:
    
    # Create a generator object for the file
    gen_file = read_large_file(file)
    
    # Print generator object
    print(gen_file)
    
    # Print the first three lines if the file
    print(next(gen_file))
    print(next(gen_file))
    print(next(gen_file))
```

### * Using the generator function to load data in chunks

* We will process the file line by line, to create a dictionary of the counts of how many times each `country` appears in a column in the dataset. 
* This time we will process the entire dataset instead of just 1000 rows of data.

```python
# Open connection to file 'world_dev_ind.csv'
with open('world_dev_ind.csv') as file:
    
    # Skipe the column names
    file.readline()
    
    # Initialize an empty dictionary
    counts_dict = {}
    
    # instead of 1000 rows we process the entire dataset
    # using the generator function
    for line in read_large_file(file):
        
        # Split the current line into a list
        row = line.split(',')
        
        # Get the value for the first column
        first_col = row[0]
        
        # If the column value is in the dict, increment its value
        if first_col in counts_dict.keys():
            counts_dict[first_col] += 1
        else:
            counts_dict[first_col] = 1
            
for key, value in counts_dict.items():
    print("{} : {} times".format(key,value))
```

### * Writing an iterator to load data in chunks

* We can use `pd.read_csv(filename, chunksize=1000)` which creates an `iterable reader object`, which means that you can use `next()` on it
* Using `'ind_pop_data.csv'` we will create another DataFrame composed of only the rows from a specific `Country Code: 'CEB'`

```python
# Initialize reader object: urb_pop_reader
urb_pop_reader = pd.read_csv('ind_pop_data.csv', chunksize=1000)

# Get the first DataFrame chunk: df_urb_pop
df_urb_pop = next(urb_pop_reader)

# Check out the head of the DataFrame
print(df_urb_pop.head())

# Check out specific country: df_pop_ceb
df_pop_ceb = df_urb_pop[df_urb_pop['CountryCode'] == 'CEB']

# Zip DataFrame columns of interest: pops
pops = zip(df_pop_ceb['Total Population'], df_pop_ceb['Urban population (% of total)'])

# Turn zip object into list: pops_list
pops_list = list(pops)
```

### * Creating new column using List Comprehensions

* We will be using a `list comprehension` to create the values for a new column `'Total Urban Population'`
* The values of this new column are the product of the values from the columns `'Total Population'` and `'Urban population (% of total)'`
* Because one of these values is a percentage, we need to multiply it by `0.01`
* We will also plot the data from this new column to creat a `visualization` of the `urban population data` using `matplotlib` (already imported as `plt`)
* The plot will be a scatter plot where the `x-axis` are values from the `'Year'` column and the `y-axis` are values from the `'Total Urban Population'` column.

```python
# Initialize reader object: urb_pop_reader
urb_pop_reader = pd.read_csv('ind_pop_data.csv', chunksize=1000)

# Get the first DataFrame chunk: df_urb_pop
df_urb_pop = next(urb_pop_reader)

# Check out the head of the DataFrame
print(df_urb_pop.head())

# Check out specific country: df_pop_ceb
df_pop_ceb = df_urb_pop[df_urb_pop['CountryCode'] == 'CEB']

# Zip DataFrame columns of interest: pops
pops = zip(df_pop_ceb['Total Population'], df_pop_ceb['Urban population (% of total)'])

# Turn zip object into list: pops_list
pops_list = list(pops)

# Use List comprehension to create new DataFrame column 
# 'Total Urban Population'
df_pop_ceb['Total Urban Population'] = [int(tup[0] * tup[1] * 0.01) for tup in pops_list]

# Plot urban population data
df_pop_ceb.plot(kind='scatter', x='Year', y='Total Urban Population')
plt.show()
```

### * Processing entired Dataset in Chunks

### * Putting it all into one Function

* We'll define a function that takes two arguments: the **`filename`** of the file to be processed, and the **`country code`** of the rows you want to process in the dataset

* Based on what was previously done, this function already does the following:
  * Loading of the file chunk by chunk
  * Creating the new column of urban population values
  * Plotting the urban population data


```python
def plot_pop(filename, country_code):
    
    # Initialize reader object
    urb_pop_reader = pd.read_csv(filename, chunksize=1000)
    
    # Initialize empty DataFrame
    data = pd.DataFrame()

    # Iterate over each DataFrame chunk
    # Chunk = 'df_urb_pop'
    for df_urb_pop in urb_pop_reader:

        # Chek out specific country
        df_pop_ceb = df_urb_pop[df_urb_pop['CountryCode'] == country_code]

        # Zip DataFram columns of interest
        pops = zip(df_pop_ceb['Total Population'], df_pop_ceb['Urban population (% of total)'])

        # Turn zip object into a list
        pops_list = list(pops)

        # Use list comprehension to create new DataFrame column
        # 'Total Urban Population'
        df_pop_ceb['Total Urban Population'] = [int(tup[0] * (tup[1] * 0.01)) for tup in pops_list]

        # Append DataFrame chunk to data
        data = data.append(df_pop_ceb)

    # Plot urban population data
    data.plot(kind='scatter', x='Year', y='Total Urban Population')
    plt.show()
```

```python
# Set the filename: fn
fn = 'ind_pop_data.csv'

# Call plot_pop for country code 'CEB'
plot_pop(fn, 'CEB')
```
